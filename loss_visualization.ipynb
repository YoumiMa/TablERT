```python
import csv
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def read_loss(dir_):
    
    train_loss_f = os.path.join(dir_, 'loss_train.csv')
    train_avg_loss_f = os.path.join(dir_, 'loss_avg_train.csv')
    val_f = os.path.join(dir_, 'eval_valid.csv')
    
    train_loss_df = pd.read_csv(train_loss_f, header=0, delimiter=';')
    train_avg_loss_df = pd.read_csv(train_avg_loss_f, header=0, delimiter=';')
    val_df = pd.read_csv(val_f, header=0, delimiter=';')
    
    return train_loss_df, train_avg_loss_df, val_df
    
```

```python
train_loss, train_avg_loss, val_df = read_loss('data/log_toy/conll04_tf_ner_10/2020-02-19_14:21:32.116657/')
```
