```python
import csv
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def read_loss(dir_):
    
    train_loss_f = os.path.join(dir_, 'loss_train.csv')
    train_avg_loss_f = os.path.join(dir_, 'loss_avg_train.csv')
    val_f = os.path.join(dir_, 'eval_valid.csv')
    
    train_loss_df = pd.read_csv(train_loss_f, header=0, delimiter=';')
    train_avg_loss_df = pd.read_csv(train_avg_loss_f, header=0, delimiter=';')
    val_df = pd.read_csv(val_f, header=0, delimiter=';')
    
    return train_loss_df, train_avg_loss_df, val_df
    
```

```python
train_loss, train_avg_loss, val_df = read_loss('data/log/conll04_tf_50_cosine_wu_rs/2020-03-03_15:18:49.117868')
```

```python
loss = pd.DataFrame([eval(t) for t in train_avg_loss['loss_avg']], columns=['total_loss', 'entity_loss', 'rel_loss'])

del train_avg_loss['loss_avg']

train_avg_loss = pd.concat([train_avg_loss, loss], axis=1)
```

```python
loss = pd.DataFrame([eval(t) for t in val_df['loss']], columns=['total_loss', 'entity_loss', 'rel_loss'])

del val_df['loss']

val_df = pd.concat([val_df, loss], axis=1)
```
